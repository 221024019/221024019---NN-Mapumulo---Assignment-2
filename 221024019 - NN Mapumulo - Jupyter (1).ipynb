```python
 Rating_column = 'Rating'

def geometric_mean(Rating_data):
    Rating_product = np.log(Rating_data)
    geometric_mean = np.exp(np.mean(Rating_product))
    return geometric_mean

geometric_mean_value = geometric_mean(Dataset[Rating_column])

print("The Geometric mean is:", geometric_mean_value)
```

## Empirical Formula based on Rating 

```python

Rating_column = 'Rating'
def empirical_rule(data):
    mean = np.mean(data)
    std_deviation = np.std(data)
    
    
    within_one_std = np.mean(np.abs(data - mean) <= std_deviation)
    within_two_std = np.mean(np.abs(data - mean) <= 2 * std_deviation)
    within_three_std = np.mean(np.abs(data - mean) <= 3 * std_deviation)
    
    return within_one_std, within_two_std, within_three_std

proportion_within_one_std, proportion_within_two_std, proportion_within_three_std = empirical_rule(Dataset[Rating_column])

print("Proportion within one standard deviation:", proportion_within_one_std)
print("Proportion within two standard deviations:", proportion_within_two_std)
print("Proportion within three standard deviations:", proportion_within_three_std)
```

## Empirical formula shape based on Rating

```python
 
Rating_column = 'Rating'
def plot_empirical_rule_curve(data):
     mean = np.mean(data)
     std_deviation = np.std(data)
     plt.figure(figsize=(10, 6))
     sns.kdeplot(data, color='skyblue', fill=True)
    
    
     for i in range(1, 4):
         plt.axvline(x=mean + i * std_deviation, color='r', linestyle='--', linewidth=2)
         plt.axvline(x=mean - i * std_deviation, color='r', linestyle='--', linewidth=2)
    
     plt.title('Empirical Rule graph based on Ratings')
     plt.xlabel('Values')
     plt.ylabel('Density')
     plt.grid(True)
     plt.show()

plot_empirical_rule_curve(Dataset[Rating_column])
warnings.filterwarnings("ignore")
```

## Calculating the correlation coefficient bewtween Ratings and Number of Votes 

```python

Rating_column = 'Rating'
Votes_column = 'Votes'


Rating = Dataset[Rating_column]
Votes = Dataset[Votes_column]

correlation_coefficient = np.corrcoef(Rating, Votes)[0, 1]

print("Correlation coefficient between Rating and Votes:", correlation_coefficient)
```

## Calculating the Covariance bewtween Ratings and Number of Votes

## Linear Regression based on ratings 

```python
x_column = 'Rating'
y_column = 'Votes'

x_data = Dataset[x_column]
y_data = Dataset[y_column]


coefficients = np.polyfit(x_data, y_data, 1)  
slope, intercept = coefficients
regression_line = slope * x_data + intercept

plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, color='blue', label='Data')
plt.plot(x_data, regression_line, color='red', label='Regression Line')
plt.title('Linear Regression')
plt.xlabel('Rating')
plt.ylabel('Votes')
plt.legend()
plt.grid(True)
plt.show()
```

## Plotting the pie chart based on the distribution of the top 40th Ratings for movies 

```python
exclude_columns = ['index_column','Date', 'z_scores', 'Director', 'Votes']  
column_Rating = 'Rating'

columns_to_consider = [col for col in Dataset.columns if col not in exclude_columns]

Dataset_filtered = Dataset[columns_to_consider].drop_duplicates()
top_10_values = Dataset_filtered.nlargest(40, column_Rating)
print(top_10_values.to_string(index=False))

#Pie chart
rating_counts = top_10_values['Rating'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Top Ratings')

plt.axis('equal') 
plt.show()

```

## Plotting the pie chart based on the distribution of the low 40th Ratings for movies

```python
exclude_columns = ['index_column','Date', 'z_scores', 'Director', 'Votes']  
column_Rating = 'Rating'

columns_to_consider = [col for col in Dataset.columns if col not in exclude_columns]

Dataset_filtered = Dataset[columns_to_consider].drop_duplicates()
top_10_values = Dataset_filtered.nsmallest(40, column_Rating)
print(top_10_values.to_string(index=False))

#Pie chart
rating_counts = top_10_values['Rating'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of low Ratings')

plt.axis('equal') 
plt.show()

```
